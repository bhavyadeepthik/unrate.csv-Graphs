## Superimposing curves


- Set the display area to 6 inches wide and 3 inches high.
- Generate 2 curves in a single graph using the 'MONTH' column for the x-axis instead of the 'DATE' column:
>- One curve with 1948 values in 'red' color.
>- A curve with the 1949 values of color 'blue'.
- Use plt.show() to display the curves.


```python
Year = unrate['DATE'].dt.year
unrate['Year'] = Year
unrate['Year']
Month = unrate['DATE'].dt.month
unrate['Month'] = Month
```

```python
MONTH =  unrate['Month'].iloc[:12]
MONTH_1=  unrate['Month'].iloc[12:24]
MONTH_2=  unrate['Month'].iloc[24:36]
MONTH_3=  unrate['Month'].iloc[36:48]
MONTH_4=  unrate['Month'].iloc[48:60]
```

```python
f = plt.figure(figsize=(6,3))
plt.plot(MONTH, UNRATE, label='1948',color='red')
plt.plot(MONTH_1, UNRATE_1, label='1949',color='blue')
plt.legend()
plt.tight_layout()
plt.show()
```
